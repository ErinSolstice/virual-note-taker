```python
import csv
import glob
import itertools
import random
import os

from trdg.generators import GeneratorFromDict
```

```python
output_dirs = ["bel_train", "bel_val_1", "bel_val_2", "bel_test"]
numImgs = [3000, 500, 500, 1000]

for (output_dir, num) in zip(output_dirs, numImgs):
    generator = GeneratorFromDict(
        fonts=glob.glob("fonts/*"),
        count=num,
        language="en",
        length=random.randint(1,2),
        width=200,
        is_handwritten=True,
        
        size=64,
        skewing_angle=15,
        random_skew=True,
        text_color="#000000,#ffffff"
    )
    
    os.makedirs(output_dir, exist_ok=True)
    with open(os.path.join(output_dir, "labels.csv"), 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        # write csv header
        writer.writerow(["filename", "words"])
        for i, (img, label) in enumerate(generator):
            img.save(f"{output_dir}/{i}.jpg")
            writer.writerow([f"{i}.jpg", label])
```
